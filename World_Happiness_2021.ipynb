   # Exploratory Data Analysis of World Happiness Report

The **World Happiness Report** is a publication of the **United Nations Sustainable Development Solutions Network**. It contains articles and rankings of national happiness, based on respondent ratings of their own lives, which the report also correlates with various (quality of) life factors. 

### The Methodology <br>

The rankings of national happiness are based on a Cantril ladder survey. Nationally representative samples of respondents are asked to think of a ladder, with the best possible life for them being a 10, and the worst possible life being a 0. They are then asked to rate their own current lives on that 0 to 10 scale. <br>Each country is also compared against a hypothetical nation called Dystopia. Dystopia represents the lowest national averages for each key variable and is, along with residual error, used as a regression benchmark.


### Importance <br>
World Happiness Report offers an insight into the progress that has been made  both in terms of understanding the determinants of global wellbeing and in recognising the most important challenges and opportunities that lie ahead.

### Dealing with Missing Values

```python
df.info()
```

In this we actually didn't have any null values, so we don't need to deal with missing values.
But, if we do find one we need to deal with **missing values** as it would severely hamper our Analysis

So what I did is, I went ahead and deleted 1 value each from Happiness_score and Trust_Government_Corruption and dealt with those by replacing the null values with the median.

<ul>
<li>Drop the missing values  If the dataset is huge and missing values are very few then we can directly drop the values because it will not have much impact.</li>
<li>Replace with mean values  We can replace the missing values with mean values, but this is not advisable in case if the data has outliers.</li>
<li>Replace with median values  We can replace the missing values with median values, and it is recommended in case if the data has outliers.</li>
<li>Replace with mode values  We can do this in the case of a Categorical feature.</li>
</ul>

```python
median_happiness_score = df['Happiness_Score'].median()
median_freedom_score = df['Trust_Government_Corruption'].median()
```

```python
df["Happiness_Score"].replace(np.nan, median_happiness_score, inplace=True)
df["Trust_Government_Corruption"].replace(np.nan, median_freedom_score, inplace=True)
```

```python
df.isnull().sum()
```

> This confirms that now our dataset is free from Null Values !!!

## Now, Let's see some Regional Analysis

### What are the happiest regions of the world in 2021?

```python
region_lists=list(df['Region'].unique())
region_happiness_ratio=[]
for each in region_lists:
    region=df[df['Region']==each]
    region_happiness_rate=sum(region.Happiness_Score)/len(region)
    region_happiness_ratio.append(region_happiness_rate)
    
data=pd.DataFrame({'region':region_lists,
                   'region_happiness_ratio':region_happiness_ratio})

new_index=(data['region_happiness_ratio'].sort_values(ascending=False)).index.values
data_region = data.reindex(new_index)

data_region
```

```python
# Create a new figure of size (12,10)
plt.figure(figsize=(12,10))
sns.barplot(x=data_region['region'], y=data_region['region_happiness_ratio'],
            palette=sns.cubehelix_palette(reverse = True, n_colors = len(data_region['region'])))
# Place the region names at a 90-degree angle.
plt.xticks(rotation= 90)
plt.xlabel('Region')
plt.ylabel('Region Happiness Ratio')
plt.title('Happiness rate for regions')
plt.show()
```

   #### North America and ANZ has the highest region wise happiness score

### How does the contribution of Economy toward's Happiness differ across various regions ?

```python
region_lists = list(df['Region'].unique())
region_economy_ratio = []

for each in region_lists:
    region = df[df['Region'] == each]
    region_economy_rate = sum(region.Economy_GDP)/len(region)
    region_economy_ratio.append(region_economy_rate)
    
data = pd.DataFrame({'region': region_lists,
                    'region_economy_ratio': region_economy_ratio})
new_index = (data['region_economy_ratio'].sort_values(ascending=True)).index.values
data_economy = data.reindex(new_index)
data_economy
```

```python
plt.figure(figsize=(15,8))

sns.barplot(x=data_economy["region"], y = data_economy["region_economy_ratio"],
            palette = "rocket")

plt.title("Economy Rate for Regions")
plt.xlabel("Region")
plt.ylabel("Economy (GDP)")
plt.xticks(rotation=90)
plt.show()
```

>**Western Europe** is the region where the contribution of **Economy (GDP Per Capita)** toward's Happiness is **highest**, >which is quite explainable due to th fact that western Europe comprises **9/10** of the **world's Happiest** countries.

### What is the ratio of the factors affecting the Happiness Score according to the regions?

```python
region_lists = list(df['Region'].unique())
economy_factor = []
SocialSupport_factor = []
Health_factor = []
Freedom_factor = []
Trust_factor = []

for each in region_lists:
    region = df[df['Region'] == each]
    economy_factor.append(sum(region.Economy_GDP)/len(region))
    SocialSupport_factor.append(sum(region.Social_Support)/len(region))
    Health_factor.append(sum(region.Health_life_expectancy)/len(region))
    Freedom_factor.append(sum(region.Freedom)/len(region))
    Trust_factor.append(sum(region.Trust_Government_Corruption)/len(region))
```

```python
fig,ax = plt.subplots(figsize=(9,5))

sns.set_color_codes("pastel")

sns.barplot(x = economy_factor, y = region_lists, color = 'g', label = 'Economy')
sns.barplot(x = SocialSupport_factor, y = region_lists, color = 'b', label = 'Family')
sns.barplot(x = Health_factor, y = region_lists, color = 'y', label = 'Health')
sns.barplot(x = Freedom_factor, y = region_lists, color = 'r', label = 'Freedom')
sns.barplot(x = Trust_factor, y = region_lists, color = 'c', label = 'Trust')

ax.legend(loc = "lower right")
ax.set(xlabel = "Proportion of various factors", ylabel = "Region", title = "Factors Affecting Happiness Score")
plt.show()
```

## Heatmap

We can find the pairwise **correlation** between the different **columns** of the data using the **corr()** method.<br>
All non-numeric data type columns are ignored here.

df.corr() is used to find the pairwise correlation of all columns in the data frame. Any nan values are automatically excluded.

The resulting coefficient is a value between -1 and 1 inclusive, where:

<ul>
<li>1: Total positive linear correlation</li>
<li>0: No linear correlation, the two variables most likely do not affect each other</li>
<li>-1: Total negative linear correlation</li>
</ul>

Now, a **heatmap** is created to visualize the correlation between the different columns of our data.

```python
plt.figure(figsize=(12,7))
sns.heatmap(df.corr(), annot = True, cmap = "RdYlGn")
plt.show()
```

As we can observe from the above heatmap of correlations, there is a **high correlation** between 

<ul>
<li>Economy (GDP per Capita)  Health (Life Expectancy) = 0.86</li>
<li>Happiness Score  Economy (GDP per Capita) = 0.79</li>
<li>Happiness Score  Health (Life Expectancy) = 0.77</li>
